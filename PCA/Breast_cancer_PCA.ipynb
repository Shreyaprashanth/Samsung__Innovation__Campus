```python
import sklearn

from sklearn.datasets import load_breast_cancer
```

```python
df=load_breast_cancer()
```

```python
X=df.data
y=df.target
```

```python
X,y
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_breast_cancer = scaler.fit_transform(X)
```

```python
pca= PCA()
pca.fit(scaled_breast_cancer)
```

```python
explained_variance_breast_cancer = pca.explained_variance_ratio_
```

```python
explained_variance_breast_cancer
```

```python
print(f"Explained variance by each component (iris): {explained_variance_breast_cancer}")
```

```python
pca.components_
```

```python
import pandas as pd
```

```python
components_breast_cancer = pd.DataFrame(pca.components_,
                               columns=df.feature_names)
```

```python
print("Principal components (loadings) for iris:")
print(components_breast_cancer)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_classifier = RandomForestClassifier()
```

```python
rf_classifier.fit(X, y)
```

```python
importances_breast_cancer = rf_classifier.feature_importances_
```

```python
feature_importance_df = pd.DataFrame({
    'Feature': df.feature_names,
    'Importance': importances_breast_cancer
}).sort_values(by='Importance', ascending=False)
```

```python
print("Feature importances for iris:")
print(feature_importance_df)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
rf.fit(X, y)
```

```python
importances = rf.feature_importances_

# Creating a DataFrame for feature importances
feature_importance = pd.DataFrame({
    'Feature': df.feature_names,
    'Importance': importances}).sort_values(by='Importance', ascending=False)

print("Feature importances for California Housing:")
print(feature_importance)
```
