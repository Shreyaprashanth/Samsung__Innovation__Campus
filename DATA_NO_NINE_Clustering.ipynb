```python
data='/content/drive/MyDrive/0.Samsung CSR AIML/Resources/SIC-Artificial-Intelligence/SIC_AI_Coding_Exercises/SIC_AI_Chapter_05_Coding_Exercises/data_number_nine.csv'
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv(data)

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
cluster = KMeans(n_clusters=3, random_state=0)

cluster.fit(df)
```

```python
cluster.fit_predict(df)
```

```python
df['cluster'] = cluster.fit_predict(df)
df.head()
```

```python
df['cluster'].value_counts()
```

```python
df['cluster'].value_counts().hist()
```

```python
import seaborn as sns

sns.pairplot(df, hue='cluster')
```

```python
sns.boxplot(x=df['X22'],y=df['cluster'])
```

```python
sns.heatmap(df.corr(),annot=True)
```
